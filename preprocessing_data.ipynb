```python
import pandas as pd
import numpy as np
from config.config import OUTPUT_CSV_FILE

df = pd.read_csv(OUTPUT_CSV_FILE)
df
```

```python
data = df
data = data.drop(columns=["Unnamed: 0", "item_id"])
data
```

```python
column_mapping = {
    'Şəhər': 'city',
    'Marka': 'marka',
    'Model': 'model',
    'Buraxılış ili': 'year',
    'Ban növü': 'type',
    'Rəng': 'color',
    'Mühərrik': 'motor',
    'Yürüş': 'distance'
}

data.rename(columns=column_mapping, inplace=True)
data
```

```python
data['motor_volume'] = data['motor'].str.extract(r'([\d.]+)\s*L')
data['horse_power'] = data['motor'].str.extract(r'(\d+)\s*a\.g\.')
data['motor_type'] = data['motor'].str.extract(r'/\s*[^/]+/\s*(.+)')

data['motor_volume'] = data['motor_volume'].astype(float)
data['horse_power'] = data['horse_power'].astype(int)

data = data.drop(columns=["motor"])
data
```

```python
data['distance'] = data['distance'].str.replace('km', '', regex=False).str.replace(' ', '').astype(int)
data
```

```python
data.loc[data['currency'] == 'USD', 'amount'] *= 1.7
data = data.drop(columns=["currency"])
data
```
