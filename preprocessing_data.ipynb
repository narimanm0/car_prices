```python
import pandas as pd
import numpy as np
from config.config import OUTPUT_CSV_FILE

df = pd.read_csv(OUTPUT_CSV_FILE)
df
```

```python
data = df
data = data.drop(columns=["Unnamed: 0", "item_id"])
data
```

```python
column_mapping = {
    'Şəhər': 'city',
    'Marka': 'marka',
    'Model': 'model',
    'Buraxılış ili': 'year',
    'Ban növü': 'type',
    'Rəng': 'color',
    'Mühərrik': 'motor',
    'Yürüş': 'distance'
}

data.rename(columns=column_mapping, inplace=True)
data
```

```python
data['motor_volume'] = data['motor'].str.extract(r'([\d.]+)\s*L')
data['horse_power'] = data['motor'].str.extract(r'(\d+)\s*a\.g\.')
data['motor_type'] = data['motor'].str.extract(r'/\s*[^/]+/\s*(.+)')

data['motor_volume'] = data['motor_volume'].astype(float)
data['horse_power'] = data['horse_power'].astype(int)

data = data.drop(columns=["motor"])
data
```

```python
data['distance'] = data['distance'].str.replace('km', '', regex=False).str.replace(' ', '').astype(int)
data
```

```python
data.loc[data['currency'] == 'USD', 'amount'] *= 1.7
data = data.drop(columns=["currency"])
data
```

```python
data['motor_volume'] = data['motor_volume'].fillna(0)
data['motor_type'] = data['motor_type'].fillna("Elektrik")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = data.select_dtypes(include=['number']).columns.tolist()

# Detect and visualize outliers for each numeric column
for col in numeric_cols:
    # Calculate IQR
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Flag outliers
    data['is_outlier'] = (data[col] < lower) | (data[col] > upper)
    
    # Plot
    plt.figure(figsize=(8, 4))
    sns.scatterplot(x=data.index, y=col, hue='is_outlier', palette={True: 'red', False: 'blue'}, data=data)

    plt.title(f"Outliers in '{col}' using IQR Method")
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.legend(title="Outlier")
    plt.show()
```
